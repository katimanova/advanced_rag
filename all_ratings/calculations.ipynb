### Подведение всех средних результатов метрик

```python
# пути
path = '../all_ratings/eval_rag_step_1/step_1_3_1 _(1000 token)'
```

```python
import pandas as pd

df = pd.read_csv(path + '/evaluete_base_rag.csv')
df.head()
```

```python
# Столбцы, которые нужно пропустить
columns_to_skip = ['question', 'contexts', 'ground_truth', 'answer']

# Вычисление средних значений для всех остальных столбцов
mean_values = df.drop(columns=columns_to_skip).mean()

# Преобразование средних значений в словарь
mean_dict = mean_values.to_dict()

# Запись словаря в текстовый файл
with open(path + '/mean_values.txt', 'w') as f:
    for key, value in mean_dict.items():
        f.write(f"{key}: {value}\n")

print("Средние значения сохранены в файл mean_values.txt")
```
